```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
from sklearn.datasets import load_iris
iris = load_iris()
iris_X = iris.data
y = iris.target
```

```python
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(2)
iris_transformed = svd.fit_transform(iris_X)
```

```python
from sklearn import decomposition
pca = decomposition.PCA()
iris_pca = pca.fit_transform(iris_X)
```

```python
fig = plt.figure(figsize=(20,7))
ax = fig.add_subplot(121)
ax.scatter(iris_pca[:,0],iris_pca[:,1],c=y,s=40)
ax.set_title(' PCA')
ax2 = fig.add_subplot(122)
ax2.scatter(-iris_transformed[:,0],iris_transformed[:,1],c=y,s=40) #the result is -1 times different from what's shown in the book
ax2.set_title('After TruncatedSVD')
```
