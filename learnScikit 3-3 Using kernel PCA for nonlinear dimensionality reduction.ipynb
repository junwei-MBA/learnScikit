```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
from sklearn import datasets, decomposition
iris = datasets.load_iris()
iris_X = iris.data

y = iris.target
```

```python
kpca = decomposition.KernelPCA(kernel='cosine', n_components=2)
iris_X_prime = kpca.fit_transform(iris_X)
```

```python
fig = plt.figure(figsize=(20,7))
ax = fig.add_subplot(121)
ax.scatter(iris_X[:,0],iris_X[:,1],c=y,s=40)
ax.set_title('Before PCA')
ax2 = fig.add_subplot(122)
ax2.scatter(-iris_X_prime[:,0],iris_X_prime[:,1],c=y,s=40) #the result is -1 times different from what's shown in the book
ax2.set_title('After KnerlPCA')
```
